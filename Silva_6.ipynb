```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\sebas\Desktop\COP1044\Silva_6\avocado.csv')
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
filtered_df = df.query("year == 2015 & type == 'conventional' & region == 'Albany'")
selected_df = filtered_df[['Date', 'Total Bags', 'Small Bags']]
selected_df.head()
```

```python
melted_df = selected_df.melt(id_vars=['Date'], var_name='Bags', value_name='Count')
melted_df.head()
```

```python
sns.lineplot(data=melted_df, x='Date', y='Count', hue='Bags').set(title="Bag Sales Over Time in Albany (2015)", xlabel="Date", ylabel="Count")
```

```python
bins = pd.qcut(df['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
df['Sales Volume'] = bins
df.head()
```

```python
sns.countplot(data=df, x='year', hue='Sales Volume').set(title="Sales Volume Distribution by Year", xlabel="Year", ylabel="Count")
```

```python
melted_bags_df = df.melt(id_vars=['Date'], value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], var_name='Bags', value_name='Count')
melted_bags_df.head()
```

```python
melted_bags_df['year'] = melted_bags_df['Date'].dt.year
sns.relplot(kind='line', x='year', y='Count', hue='Bags', data=melted_bags_df).set_axis_labels("Year", "Count").set(title="Bag Sales by Year")
```
